```python
# Step 1: Install Required Library
!pip install pretty_midi

# Step 2: Import Required Libraries
import pretty_midi
import os
import matplotlib.pyplot as plt

# Step 3: Upload MIDI Files
from google.colab import files
uploaded = files.upload()

# Step 4: Process the Uploaded MIDI File
for filename in uploaded.keys():
    try:
        midi_data = pretty_midi.PrettyMIDI(filename)
        print(f" Processing: {filename}")

        # Print general info
        print("Number of Instruments:", len(midi_data.instruments))
        print("Duration (seconds):", midi_data.get_end_time())

        # Plot piano roll (first instrument only)
        piano_roll = midi_data.instruments[0].get_piano_roll(fs=100)
        plt.imshow(piano_roll, aspect='auto', origin='lower', cmap='hot')
        plt.title(f"Piano Roll of {filename}")
        plt.xlabel('Time')
        plt.ylabel('Pitch')
        plt.show()

    except Exception as e:
        print(f" Error reading {filename}: {e}")
```
