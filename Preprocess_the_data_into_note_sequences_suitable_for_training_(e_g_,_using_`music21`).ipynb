```python
# Step 1: Install music21
!pip install music21

# Step 2: Import Libraries
from music21 import converter, instrument, note, chord
from google.colab import files
import pickle

# Step 3: Upload MIDI Files
uploaded = files.upload()  # Upload one or multiple MIDI files

# Step 4: Extract Notes and Chords from Files
notes = []

for filename in uploaded:
    print(f"Processing {filename}...")
    try:
        midi = converter.parse(filename)

        # Flatten the MIDI file into a single stream
        parts = instrument.partitionByInstrument(midi)

        if parts:  # If instrument parts are available
            notes_to_parse = parts.parts[0].recurse()
        else:
            notes_to_parse = midi.flat.notes

        for element in notes_to_parse:
            if isinstance(element, note.Note):
                notes.append(str(element.pitch))  # e.g., "C4"
            elif isinstance(element, chord.Chord):
                notes.append('.'.join(str(n) for n in element.normalOrder))  # e.g., "60.64.67"

    except Exception as e:
        print(f" Error processing {filename}: {e}")

# Step 5: Show Sample of Collected Notes
print(f"\n Total notes/chords extracted: {len(notes)}")
print("Sample notes:", notes[:20])

# Step 6: Save the Notes for Training Later
with open('notes.pkl', 'wb') as f:
    pickle.dump(notes, f)

print("Note sequences saved as 'notes.pkl'")
```
